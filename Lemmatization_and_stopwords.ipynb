# **Lemmatization**

Lemmatization is a natural language processing (NLP) technique for reducing words to their lemma or base (dictionary) form. Unlike stemming, which simply removes suffixes (sometimes resulting in incomplete words), lemmatization uses linguistic rules and vocabulary to ensure that words are converted into their root form while retaining meaning. For example, lemmatization turns "running" into "run" and "better" into "good" by understanding the grammatical structure and context.

**Lemmatization generally requires a POS tag (e.g., noun, verb, adjective) to achieve accurate results. Without the correct POS, words may not reduce to the right form.**

```python
import nltk
nltk.download('wordnet')
```

```python

from nltk.stem import WordNetLemmatizer

'''

POS tag -

Noun-n
verb-V I
adjective-a
adverb-r

'''
# by default pos tag is 'n'

Lemmatizer = WordNetLemmatizer()

Lemmatizer.lemmatize("going", pos = 'v')
```

```python

words = ["eating", "eats", "eat", "ate", "adjustable", "rafting", "ability", "meeting"]

for word in words:
  print(word, " ---> ", Lemmatizer.lemmatize(word , pos = 'v'))
```

# **STOPWORDS**

In natural language processing (NLP), stopwords are common words that are often removed from text data before processing, as they carry little meaningful information for analysis. These words typically include high-frequency terms like "the," "is," "in," "and," and "to." While they are essential for grammatical structure, they don’t usually add significant value for tasks like text classification, sentiment analysis, or topic modeling.

Why Remove Stopwords?
Reduce Noise: Stopwords can dilute the impact of more meaningful words, so removing them helps focus on the keywords that actually contribute to the content’s meaning.
Enhance Efficiency: Text data processing becomes faster and more memory-efficient when redundant words are removed, especially in large datasets.
Improves Accuracy: By removing stopwords, the algorithms focus on more significant words, which can improve the accuracy of NLP models in tasks like search and classification.
Common Stopwords Lists
Libraries like NLTK, SpaCy, and Scikit-learn come with pre-defined stopwords lists for various languages. However, these lists can be customized to fit the specific needs of an application, as sometimes words in the standard stopwords lists might be relevant in a specific context.

```python
# Now let's apply stopwords and stemming
import nltk
nltk.download('punkt')

from nltk.stem import SnowballStemmer

snowballstemmer = SnowballStemmer('english')

sentences = nltk.sent_tokenize(paragraph)
```

```python
# now we have a list of sentences on which we will apply stopwords.
# the words that are not present in the set of stopwords module, we will take those words and apply stemming on them.

for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  words = [snowballstemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
  sentences[i] = ' '.join(words) # join() will combine all the words to form a sentence


# we are using set bcz some of the words may get repeated.
```
